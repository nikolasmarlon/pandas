# Séries

```python
import pandas as pd


lista_numeros = [39, 12, 343, 34, 43, 10]

# Series do pandas
serie = pd.Series(lista_numeros) # Retorna índice e valor numa coluna
print(serie)


# Acessar pelo índice
print(serie[2])
```

```python
# Criar serie com índice personalizado
index=['a', 'b', 'c', 'd', 'e', 'f']
serie_2 = pd.Series(lista_numeros, index)
print(serie_2)

# Acessar pelo índice personalizado
print(serie_2['c'])
```

```python
# Serie a partir de um dicionário

# Dicionário de números
dicionario_numeros = {'a': 1, 'b': 3, 'c': 5, 'd': 7, 'e': 9}
serie_3 = pd.Series(dicionario_numeros)
print(serie_3)
```

# Dataframes

# Head, Tail, Shape, Info, Describe
Funções do pandas

```python
import pandas as pd

# Dicionário
dados = {
    'Nome': ['João', 'Maria', 'Pedro', 'Ana'],
    'Idade': [25, 49, 18, 43],
    'Cidade': ['São Paulo', 'Rio de Janeiro', 'Belo Horizonte', 'Salvador'],
    'Salário': [5000, 7000, 3000, 6000]
}

df = pd.DataFrame(dados)
print(df)


print('\n')
```

```python
# Função head(n) - vai retornar as primeiras linhas do dataframe
print(df.head(2)) # Padão retorna 5 linhas se não for informado
```

```python
# Função tail() vai retornar as últimas linhas
print(df.tail(2))
```

```python
# Função shape()
print(df.shape) # Retorna (linhas, colunas)
```

```python
# Função info()
print(df.info()) # Retorna informações do dataframe
```

```python
# Função describe()
print(df.describe()) # Retorna informações estatísticas do dataframe ( estatística descritiva)
```

# Loc e Iloc
Seleção de dados

```python
import pandas as pd

#Criar dataframe DF
dados = {
    'Nome': ['João', 'Maria', 'Pedro', 'Ana'],
    'Idade': [25, 34, 21, 44],
    'Cidade': ['São Paulo', 'Rio de Janeiro', 'Belo Horizonte', 'Salvador']
}

# Criar index
index = ['a', 'b', 'c', 'd']

df = pd.DataFrame(dados, index) # Dataframe com índice personalizado
print(df)


# Visualizar colunas específicas
print('\n')
#df.Cidade # Forma 1
df['Cidade'] # Forma 2

```

```python
# Selecionar multiplas colunas
df[['Nome', 'Cidade']]
```

```python
# loc - usando rótulos
df.loc['a'] # Retorna a linha com o rótulo 'a'

print('\n')
# Retornando coluna
df.loc[['a', 'b'], ['Nome']] # Retorna o valor da linha 'a' e coluna 'Nome'
```

```python
# iloc - index baseado no índice numéricos
df.iloc[0] # Retorna a primeira linha
```

```python
df.iloc[0:2] # Retorna as duas primeiras linhas
```

```python
# Selecionar linhas 1 e 2 e coluna 1
df.iloc[0:2, 0]
```

# Filtrando dados

```python
# Dados para o dataframe - nome, idade, cidade, score
dados = {
    'Nome': ['João', 'Maria', 'Pedro', 'Ana', 'Pedro'],
    'Idade': [25, 34, 21, 23, 41],
    'Cidade': ['São Paulo', 'Rio de Janeiro', 'Belo Horizonte', 'Salvador', 'Rio de Janeiro'],
    'Score': [85, 90, 75, 80, 93]
}

# Dataframe
df = pd.DataFrame(dados)
print(df)
```

```python
# Filtrar pessoas com score maores que 80
df[df['Score'] > 80]
```

```python
# Filtar pessoas do RJ com score maior ou igual a 90
df[(df['Cidade'] == 'Rio de Janeiro') & (df['Score'] >= 90)]
```

```python
# Filtrar por pessoas do rio ou são paulo
df[(df['Cidade'] == 'Rio de Janeiro') | (df['Cidade'] == 'São Paulo')]
```

```python
# Método isin()
df[df['Cidade'].isin(['Rio de Janeiro', 'São Paulo'])]
```

# Sort values - ordenar crescente ou decrescente

```python
import pandas as pd

dados_frutas = {
    'Fruta': ['Maçã', 'Banana', 'Laranja', 'Morango', 'Uva', 'Abacaxi'],
    'Estoque': [100, 150, 80, 200, 120, 50],
    'Preço': [2.50, 1.80, 3.20, 4.00, 5.50, 7.00]
}

df_frutas = pd.DataFrame(dados_frutas)
print(df_frutas)
```

```python
# Ordenar em ordem crescente
df_frutas.sort_values(by='Estoque')
```

```python
# Ordenar por preço
df_frutas.sort_values(by='Preço')
```

```python
# Ordenar em ordem alfabética
df_frutas.sort_values(by='Fruta')
```

```python
# Ordenar em ordem decrescente por preço
df_frutas.sort_values(by='Preço', ascending=False)
```

```python
# Ordenar em ordem alfabética decrescente
df_frutas.sort_values(by='Fruta', ascending=False)
```

```python
# Ordenação múltiplas colunas.
df_frutas.sort_values(by=['Estoque', 'Preço'], ascending=[True, False])
```
