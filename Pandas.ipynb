# Séries

```python
import pandas as pd


lista_numeros = [39, 12, 343, 34, 43, 10]

# Series do pandas
serie = pd.Series(lista_numeros) # Retorna índice e valor numa coluna
print(serie)


# Acessar pelo índice
print(serie[2])
```

```python
# Criar serie com índice personalizado
index=['a', 'b', 'c', 'd', 'e', 'f']
serie_2 = pd.Series(lista_numeros, index)
print(serie_2)

# Acessar pelo índice personalizado
print(serie_2['c'])
```

```python
# Serie a partir de um dicionário

# Dicionário de números
dicionario_numeros = {'a': 1, 'b': 3, 'c': 5, 'd': 7, 'e': 9}
serie_3 = pd.Series(dicionario_numeros)
print(serie_3)
```

# Filtrando dados

```python
# Dados para o dataframe - nome, idade, cidade, score
dados = {
    'Nome': ['João', 'Maria', 'Pedro', 'Ana', 'Pedro'],
    'Idade': [25, 34, 21, 23, 41],
    'Cidade': ['São Paulo', 'Rio de Janeiro', 'Belo Horizonte', 'Salvador', 'Rio de Janeiro'],
    'Score': [85, 90, 75, 80, 93]
}

# Dataframe
df = pd.DataFrame(dados)
print(df)
```

```python
# Filtrar pessoas com score maores que 80
df[df['Score'] > 80]
```

```python
# Filtar pessoas do RJ com score maior ou igual a 90
df[(df['Cidade'] == 'Rio de Janeiro') & (df['Score'] >= 90)]
```

```python
# Filtrar por pessoas do rio ou são paulo
df[(df['Cidade'] == 'Rio de Janeiro') | (df['Cidade'] == 'São Paulo')]
```

```python
# Método isin()
df[df['Cidade'].isin(['Rio de Janeiro', 'São Paulo'])]
```

# Sort values - ordenar crescente ou decrescente

```python
import pandas as pd

dados_frutas = {
    'Fruta': ['Maçã', 'Banana', 'Laranja', 'Morango', 'Uva', 'Abacaxi'],
    'Estoque': [100, 150, 80, 200, 120, 50],
    'Preço': [2.50, 1.80, 3.20, 4.00, 5.50, 7.00]
}

df_frutas = pd.DataFrame(dados_frutas)
print(df_frutas)
```

```python
# Ordenar em ordem crescente
df_frutas.sort_values(by='Estoque')
```

```python
# Ordenar por preço
df_frutas.sort_values(by='Preço')
```

```python
# Ordenar em ordem alfabética
df_frutas.sort_values(by='Fruta')
```

```python
# Ordenar em ordem decrescente por preço
df_frutas.sort_values(by='Preço', ascending=False)
```

```python
# Ordenar em ordem alfabética decrescente
df_frutas.sort_values(by='Fruta', ascending=False)
```

```python
# Ordenação múltiplas colunas.
df_frutas.sort_values(by=['Estoque', 'Preço'], ascending=[True, False])
```

# Tratamento de dados - Astype, Isna, Fillna e Dropna

```python
# Criar dataframe
import pandas as pd
import numpy as np # bom para trabalhar com dados ausentes

dados = {
    'Nome': ['João', 'Maria', 'Pedro', 'Ana'],
    'Idade': ['25', '34', '21', '23'],
    'Salário': [5000.75, 1700.20, 3000.99, 6000],
    'Cidade': ['São Paulo', 'Rio de Janeiro', np.nan, 'Salvador']
}

df = pd.DataFrame(dados)
print(df)
```

```python
# pandas entende string como objeto( object )
# Ferificar tipo
df.info()
```

```python
# Converter idade 'string' para 'int'
df.Idade = df.Idade.astype(int)

df.info()
```

```python
print(df)
```

```python
# Converter salário para int -- float para int -- obs. não arredonda. pega a parte inteira
df.Salário = df.Salário.astype(int)
print(df)
```

```python
df.info()
```

```python
# Valores ausentes
# Retornar um dataframe com valores booleanos com isna()
print(df.isna())
```

```python
# fillna() permite prencher algum valor ausente com algum valor

# ex. preencher com 0 onde tem valores nulos
df_nulos = df.fillna({'Cidade': 'Não informado'}) # passar um dicionário
print(df_nulos)
```

```python
# Dataframe com valores de salário ausentes do tipo string
dados_ausentes = {
    'Nome': ['João', 'Maria', 'Pedro', 'Ana'],
    'Idade': [25, 34, 21, 23],
    'Salário': [5000.75, 1700.20, np.nan, 6000],
    'Desconto': [0.3, 0.44, 0.88, np.nan]
}

df_ausentes = pd.DataFrame(dados_ausentes)
print(df_ausentes)
```

```python
# Preencher salário com média de valores da coluna
media_salario = df_ausentes.fillna({'Salário': df_ausentes.Salário.mean()})
print(media_salario)
```

```python
# Preencher salário com mediana dos valores da coluna
mediana_salario = df_ausentes.fillna({'Desconto': df_ausentes.Desconto.median()})
print(mediana_salario)
```

```python
# Remover linhas
df_drop_linhas = df_ausentes.dropna()
print(df_drop_linhas)
```

```python
# Remover colunhas
df_drop_colunas = df_ausentes.dropna(axis=1) # axis=1 é coluna
print(df_drop_colunas)
```

# aplicar função em eixos

```python
import pandas as pd
dados = {
    'Nome': ['João', 'Maria', 'Pedro', 'Ana'],
    'Idade': [25, 34, 21, 23],
    'Status': ['Ativo', 'Inativo', 'Ativo', 'Inativo']
}

df = pd.DataFrame(dados)
print(df)
```

```python
# apply  - criar função para verificar se a pessoa é jovem ou adulto

def categoria_idade ( idade ):
  if idade < 30:
    return 'Jovem'
  else:
    return 'Adulto'

df['Categoria_Idade'] = df['Idade'].apply(categoria_idade)
print(df)
```

```python
# Map


# Dicionário
mapeamento_status = { 'Ativo': 'ATIVO', 'Inativo': 'INATIVO'}

# Aplicar mapeamento
df['Status_Mapeado'] = df['Status'].map(mapeamento_status)
print(df)
```

```python
# Apply com função lambda
# Função para dobrar a idade
df['Dobro_Idade'] = df['Idade'].apply(lambda x: x * 2)
print(df)
```

# Funções de agregação
serve para resumir e analisar, agrupar dados em uma coluna, ... agrupar por cidade, agrupar por cidade.

```python
import pandas as pd

dados = {
    'Nome': ['João', 'Maria', 'Pedro', 'Ana', 'João', 'Maria'],
    'Cidade': ['São Paulo', 'Rio de Janeiro', 'Belo Horizonte', 'Salvador', 'São Paulo', 'Rio de Janeiro'],
    'Idade': [25, 34, 2, 21, 23, 44],
    'Salário': [5000, 7000, 3000, 6000, 5000, 7000]
}

df = pd.DataFrame(dados)
print(df)
```

```python
# Agrupar por cidade e calcular a soma dos salarios
vendas_por_cidade = df.groupby('Cidade')['Salário'].sum()
print(vendas_por_cidade)
#
```

```python
# Agrupar por cidade e ver a média de idade e salários
media_idade_salario_por_cidade = df.groupby('Cidade')[['Idade', 'Salário']].mean()
print(media_idade_salario_por_cidade)
```

```python
# Dtypes - tudo que for do tipo string o pandas retorna como objeto.
print(df.dtypes)
```

```python
# converter para o formato data
df['Data'] = pd.to_datetime(df['Data'])# Convete uma coluna para o formato data
print(df.dtypes)
```

```python
# Lidar com formatos diferentes de datas
df_data_formatos = pd.DataFrame({
    'Data': ['01/02/2022', '02-Jan-2026', '20230604'],
    'Valor': [10, 20, 30]
})

print(df_data_formatos)
print(df_data_formatos.dtypes)
```

```python
# converter formatos diferentes
df_data_formatos['Data_convertida'] = pd.to_datetime(df_data_formatos['Data'], format='mixed')
print(df_data_formatos.dtypes)

print(df_data_formatos)
```

```python
# Extrair valores

df_data_formatos['Ano'] = df_data_formatos['Data_convertida'].dt.year
df_data_formatos['Mês'] = df_data_formatos['Data_convertida'].dt.month
df_data_formatos['Dia'] = df_data_formatos['Data_convertida'].dt.day
df_data_formatos['Hora'] = df_data_formatos['Data_convertida'].dt.hour
df_data_formatos['Minuto'] = df_data_formatos['Data_convertida'].dt.minute
df_data_formatos['Segundo'] = df_data_formatos['Data_convertida'].dt.second
df_data_formatos['Dia_da_semana'] = df_data_formatos['Data_convertida'].dt.day_name()

print(df_data_formatos)
#
```

# Concatenar objetos dos pandas colunas por coluna ou linhas com concat()

# Merge - combinar dataframes

```python
# criar data frame
import pandas as pd

df1 = pd.DataFrame({'chave': ['A', 'B', 'C', 'D'], 'valor1': [1, 2, 3, 5]})
df2 = pd.DataFrame({'chave': ['B', 'D', 'E', 'F'], 'valor2': [6, 7, 8, 9]})

print(df1)
print(df2)
```

```python
# Merge Inner combinar pela chave
df_merge = pd.merge(df1, df2, on='chave', how='inner') # inner é o padrão
print(df_merge)
```

```python
# left merge
merge_left = pd.merge(df1, df2, on='chave', how='left')
print(merge_left) # Quando não tem associação ele coloca o valor nulo
```

```python
# right merge
merge_right = pd.merge(df1, df2, on='chave', how='right')
print(merge_right)
```

```python
# Outer merge
merge_outer = pd.merge(df1, df2, on='chave', how='outer')
print(merge_outer)
```
